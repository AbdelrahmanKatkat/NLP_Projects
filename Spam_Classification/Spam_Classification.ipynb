# Spam Classification with SVM

## EDA

```python
df=pd.read_csv('/content/drive/MyDrive/Datasets/NLP/Spam_Text_Classification/spam.csv', encoding='latin-1')
```

```python
df.head(5)
```

```python
df.drop(columns=['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], inplace=True)
```

```python
df.rename(columns={'v1':'Class','v2':'Prompt'},inplace=True)
```

```python
df.columns
```

```python
# Sample text documents and their corresponding labels
documents = df['Prompt'].tolist()
labels = df['Class'].tolist()
```

## Preprocessing & Text Representation

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(documents, labels, test_size=0.2, random_state=42)
```

```python
# Text representation using TF-IDF
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

## Training

```python
# Train a classifier (Support Vector Machine - SVM)
classifier = LinearSVC()
classifier.fit(X_train_tfidf, y_train)
```

```python
# Predict on the test set
y_predt=classifier.predict(X_train_tfidf)
y_pred = classifier.predict(X_test_tfidf)
```

## Evaluation

```python
# Evaluate the performance
print(classification_report(y_test, y_pred))
```

```python
# Evaluate the performance
print(classification_report(y_train, y_predt))
```

```python
print(X_train[1])
print(X_train_tfidf[1])
print()
print(X_test)
print(X_test_tfidf)
```

# Spam Classification with Naive Bayes

## EDA

```python
df=pd.read_csv('/content/drive/MyDrive/Datasets/NLP/Spam_Text_Classification/spam.csv', encoding='latin-1')
```

```python
df.head(5)
```

```python
df.drop(columns=['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], inplace=True)
```

```python
df.rename(columns={'v1':'Class','v2':'Prompt'},inplace=True)
```

```python
df.columns
```

```python
# Sample text documents and their corresponding labels
documents = df['Prompt'].tolist()
labels = df['Class'].tolist()
```

## Preprocessing & Text Representation

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(documents, labels, test_size=0.2, random_state=42)
```

```python
# Text representation using CountVectorizer
vectorizer = CountVectorizer()
X_train_CV = vectorizer.fit_transform(X_train)
X_test_CV = vectorizer.transform(X_test)
```

## Training

```python
# Create a Naive Bayes classifier
classifier = MultinomialNB()
classifier.fit(X_train_CV, y_train)
```

```python
# Predict on the test set
y_predt=classifier.predict(X_train_CV)
y_pred = classifier.predict(X_test_CV)
```

## Evaluation

```python
# Evaluate the performance
print(classification_report(y_test, y_pred))
```

```python
# Evaluate the performance
print(classification_report(y_train, y_predt))
```
